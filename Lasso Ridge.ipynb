```python
import pandas as pd
```

# Load data

```python
header=[
    'symboling',
    'normalized-losses',
    'make',
    'fuel-type',
    'aspiration',
    'num-of-doors',
    'body-style',
    'drive-wheels',
    'engine-location',
    'wheel-base',
    'length',
    'width',
    'height',
    'curb-weight',
    'engine-type',
    'num-of-cylinders',
    'engine-size',
    'fuel-system',
    'bore',
    'stroke',
    'compression-ratio',
    'horsepower',
    'peak-rpm',
    'city-mpg',
    'highway-mpg',
    'price'
]
auto_data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data', 
                        sep=r'\s*,\s*', engine='python', names=header)
auto_data
```

# Scrub Data

## Replace ? with NaN

```python
import numpy as np

auto_data = auto_data.replace('?',np.nan)
auto_data.head()
```

### Describe

```python
auto_data.describe()
```

### Include all shows missing price

```python
auto_data.describe(include='all')
```

### Price is object

```python
auto_data['price'].describe()
```

#### Convert price to number
coerce forces the conversion

```python
auto_data['price'] = pd.to_numeric(auto_data['price'],errors='coerce')
```

```python
auto_data['price'].describe()
```

#### Drop normalized-losses is not used for the price

```python
auto_data=auto_data.drop('normalized-losses',axis=1)
auto_data.head()
```

#### Convert horsepower to numeric

```python
auto_data['horsepower'].describe()
```

```python
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'],errors='coerce')
```

```python
auto_data['horsepower'].describe()
```

#### Convert num-of-cylinders

```python
auto_data['num-of-cylinders'].describe()
```

```python
cylinders_dict = {
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'eight':8,
    'twelve':12
}

auto_data['num-of-cylinders'].replace(cylinders_dict, inplace=True)
auto_data.head()
```

#### One Hot of other columns 

```python
auto_data = pd.get_dummies(auto_data,
                           columns=['make',
                                    'fuel-type',
                                    'aspiration',
                                    'num-of-doors',
                                    'body-style',
                                    'drive-wheels',
                                    'engine-location',
                                    'engine-type',
                                    'fuel-system'
                                   ])
auto_data.head()
```

### Drop NA

```python
auto_data = auto_data.dropna()
auto_data
```

## Check for null

```python
auto_data[auto_data.isnull().any(axis=1)]
```

## Split data for training and testing

```python
from sklearn.model_selection import train_test_split

#Price is label
X = auto_data.drop('price', axis=1)
Y = auto_data['price']

# Split 80% for training, 20% for testing
X_train, x_test, Y_train, y_test = train_test_split(X,Y,test_size =0.2,random_state=0)
```

# Estimators

## LinearRegression

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train,Y_train)
```

### Score to get R^2  (r square)

```python
linear_model.score(X_train,Y_train)
```

### Weights for each feature

```python
linear_model.coef_
```

```python
predictors = X_train.columns
coef=pd.Series(linear_model.coef_,predictors).sort_values()
print(coef)
```

## Predict

```python
y_predict = linear_model.predict(x_test)
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (15,6)

plt.plot(y_predict,label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('price')
plt.legend()
plt.show()
```

### Lower score than predicted

```python
r_square = linear_model.score(x_test,y_test)
r_square
```

```python
from sklearn.metrics import mean_squared_error

linear_model_mse = mean_squared_error(y_predict,y_test)
linear_model_mse
```

#### Root Mean Square Error (RMSE)
    Standard deviation of residuals
    
How much will the predicted diff from the actual

```python
import math

math.sqrt(linear_model_mse)
```

## Lasso

```python
from sklearn.linear_model import Lasso

# alpha prevents overfitting of the model
# normalize centers the data by subtracting the mean and dividing by the L2 norm
# can change alpha to test
lasso_model = Lasso(alpha=.5,normalize=True)
lasso_model.fit(X_train,Y_train)
```

```python
lasso_model.score(X_train,Y_train)
```

```python
predictors = X_train.columns
coef=pd.Series(lasso_model.coef_,predictors).sort_values()
print(coef)
```

```python
y_predict = lasso_model.predict(x_test)
```

```python
pylab.rcParams['figure.figsize'] = (15,6)

plt.plot(y_predict,label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('price')
plt.legend()
plt.show()
```

Better r square than linear

```python
r_square = lasso_model.score(x_test,y_test)
r_square
```

Lower RMSE than linear

```python
lasso_model_mse = mean_squared_error(y_predict,y_test)
math.sqrt(lasso_model_mse)
```
