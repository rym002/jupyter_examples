# Show face

```python
from scipy import misc
import matplotlib
import matplotlib.pyplot as plt

image = misc.face()
plt.imshow(image)
```

## Shape
768x1024
3 channels

## Grayscale

```python
image_gray = misc.face(True)
plt.imshow(image_gray)
```

```python
image_gray.shape
```

### Add 3rd dimension to grayscale using numpy

```python
import numpy as np

expanded_grayscale = np.expand_dims(image_gray, axis = 2)
expanded_grayscale.shape
```
